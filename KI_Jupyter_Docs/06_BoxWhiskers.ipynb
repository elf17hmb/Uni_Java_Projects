# Zufällig erzeugte Daten

Daten werden zusammengesetzt aus drei gleichverteilt gezogenen Stichproben (`spread`, `flier_high`, `flier_low`). Zusätzlich werden in der Mitte des Intervalls 25 mal der 50 eingefügt (`center`).

## Aufgabe

Ändern Sie das Histogram so ab, dass sie sich jeweils die einzelnen Teilstichproben betrachten können. Halten Sie dazu den Bereich, den die x-Achse darstellt konstannt. Was passiert, wenn Sie dies auch für die jeweilgen Box-Plots machen (Halten Sie diesmal die Einteilung der y-Achse konstant)? 

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# fake up some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))

plt.hist(data, bins=30)
plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, showfliers=False)
plt.show()
```

# Darstellung realer Daten

Der ursprünglich von Keggle stammende Datensatz beinhaltet diverse Messwerte von Patienten, die Brustkrebs ausgeprägt hatten. Die Daten sind gelabelt, ob die Tumore gutartig bzw. bösartig waren. Für die Darstellung im Boxplot wird die Größe der Fläche, die der Turmor einnimmt jeweils fpr gutartige bzw. bösartige Tumore gegenübergestellt.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/'
                 + 'Python_Tutorials/master/Kaggle/BreastCancerWisconsin/data/data.csv')
```

```python
malignant = df[df['diagnosis']=='M']['area_mean']
benign = df[df['diagnosis']=='B']['area_mean']

plt.boxplot([benign, malignant], labels=['bösartig', 'gutartig'])
plt.ylabel("Größe")
plt.xlabel("Diagnose")
plt.title("Brustkrebs")
plt.show()
```

```python
df.boxplot(column = 'area_mean', by = 'diagnosis');
plt.title('OwO')
```
