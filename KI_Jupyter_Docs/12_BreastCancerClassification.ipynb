```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/'
                 + 'Python_Tutorials/master/Kaggle/BreastCancerWisconsin/data/data.csv')
df = df.drop(columns=["id","Unnamed: 32"], axis = 1)
```

```python
df.head()
```

```python
X_old = df[["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","radius_worst","texture_worst","perimeter_worst","area_worst","smoothness_worst","compactness_worst","concavity_worst","concave points_worst","symmetry_worst","fractal_dimension_worst"]].values
Y = df[["diagnosis"]].values
X=df.iloc[:,1:31].values
print(X)
print(np.array_equal(X_old,X))
print(Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import minmax_scale
from sklearn import preprocessing
from warnings import simplefilter
from helper import plot_classifier

simplefilter(action='ignore')
le = preprocessing.LabelEncoder()
Y=le.fit_transform(Y)
```

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.25)

model=LogisticRegression();
model.fit(X_train,Y_train)
print("Trainingsdaten: ", model.score(X_train,Y_train))
print("Testdaten: ", model.score(X_test,Y_test))

print(model.coef_)
print(model.intercept_)
#prediction = model.predict(X_test)
#print("prediction:",prediction)
```

```python
Xs_train = minmax_scale(X_train)
Xs_test = minmax_scale(X_test)

model=LogisticRegression();
model.fit(Xs_train,Y_train)
print("Trainingsdaten: ", model.score(Xs_train,Y_train))
print("Testdaten: ", model.score(Xs_test,Y_test))

print(model.coef_)
print(model.intercept_)
#prediction = model.predict(X_test)
#print("prediction:",prediction)
```

```python
Xn_train = preprocessing.scale(X_train)
Xn_test = preprocessing.scale(X_test)

model=LogisticRegression();
model.fit(Xn_train,Y_train)
print("Trainingsdaten: ", model.score(Xn_train,Y_train))
print("Testdaten: ", model.score(Xn_test,Y_test))

print(model.coef_)
print(model.intercept_)
#prediction = model.predict(X_test)
#print("prediction:",prediction)
```

```python
Xsn_train = preprocessing.scale(Xs_train)
Xsn_test = preprocessing.scale(Xs_test)

model=LogisticRegression();
model.fit(Xsn_train,Y_train)
print("Trainingsdaten: ", model.score(Xsn_train,Y_train))
print("Testdaten: ", model.score(Xsn_test,Y_test))

print(model.coef_)
print(model.intercept_)
#prediction = model.predict(X_test)
#print("prediction:",prediction)
```

```python
#Confusion Matrix (Accuracies)
model = LogisticRegression()
model.fit(X_train,Y_train)
y_test_pred = model.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_test_pred)
print(cm)
```

```python
plt.matshow(cm)

for (i,j), z in np.ndenumerate(cm):
    plt.text(j,i,'{:0.1f}'.format(z),
            ha='center',va='center',
            bbox=dict(boxstyle='round',
                     facecolor='white',
                     edgecolor='0.3'))
plt.colorbar()
```
