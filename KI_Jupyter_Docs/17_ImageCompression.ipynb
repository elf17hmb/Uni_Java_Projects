## Aufgabe

Clustern sie die Farben im RGB-Bild `lena_color.gif`. Verwenden Sie die gefundenen Prototypen als indizierte Farbpalette. Die im Folgenden definierte Funktion `showImage()` übernimmt dabei die Visualisierung. Der Aufruf sieht dabei wie folgt aus:

    showImage(labels, prototypes, width=512, height=512)
    
- `labels` .. Clusterzugehörigkeit jedes einzelnen Pixels
- `prototypes` .. Die Farbprototypen
- `width`& `height`.. Breite und Höhe des geclusterten Bildes (Defaultwerte beziehen sich auf `lena_color.gif`

Experimentieren Sie mit den der Anzahl Cluster und verwenden Sie auch andere Bilder. Was beobachten Sie?

```python
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread("mina_color.jpg")
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as col

def showImage(labels, prototypes, width=1200, height=1800):
    ### 
    cm=col.ListedColormap(prototypes/255.0)
    plt.figure(figsize=(20,20))
    plt.imshow(labels.reshape(1800,1200),cmap=cm)
    matfig=plt.figure(figsize=(20,2))
    plt.matshow(np.arange(0,prototypes.shape[0]).reshape(1,-1), cmap=cm, fignum=matfig.number)
```

```python
plt.imshow(img)
```

```python
from sklearn.cluster import KMeans
cluster_model = KMeans(n_clusters=5, verbose=False, n_jobs=4, n_init=2)
cluster_model.fit(img.reshape(-1,3))
```

```python
showImage(cluster_model.labels_, cluster_model.cluster_centers_)
```
