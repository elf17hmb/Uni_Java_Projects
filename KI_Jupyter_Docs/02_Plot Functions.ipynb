```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.rcParams["figure.figsize"] = (12, 5)
```

# 2D-Plot

```python
x = np.linspace(0,2*math.pi)
y1 = np.sin(x)
y2 = np.cos(x)


plt.plot(x,y1)
plt.plot(x,y2,'+')

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Plot")

plt.annotate('local max', xy=(0.5*math.pi, 1), xytext=(3, 0.9),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.show()
```

# Bar-Graph

```python
scores = [0.92, 0.95, 0.99, 0.85, 0.9992, 0.3]
index  = ['a','b','c','d','e','f']

plt.figure(figsize=(5, 3))

plt.bar(index,scores)
plt.ylim(0.25,1.05)
plt.show()
```

# Surface-Plot

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

# Stützstellen "zusammenstellen"
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

# Funktionswerte berechnen
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z)


plt.show()
```
