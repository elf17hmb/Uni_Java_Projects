```python
import matplotlib.pyplot as plt
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()
X  = iris.data[:, :2]  # we only take the first two features.
X2 = iris.data[:, 3]   # third feature also needs to be visualized
y = iris.target

print(iris.DESCR)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=y*10)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show()
```

```python
plt.scatter(X[:, 0][y==0], X[:, 1][y==0], marker='+', s=X2*100)
plt.scatter(X[:, 0][y==1], X[:, 1][y==1], marker='x', s=X2*100)
plt.scatter(X[:, 0][y==2], X[:, 1][y==2], marker='.', s=X2*100)

plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show()
```
